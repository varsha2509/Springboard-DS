### Task 3. Calculate what the highest and lowest opening prices were for the stock in this period.


```python
opening_price = []
for row in json_2017['dataset']['data']:
    opening_price.append(row[1])
```

```python
highest = max(x for x in opening_price if x is not None)
lowest = min(x for x in opening_price if x is not None)
```

```python
print("Maximum opening price is ${} and the minimum opening price is ${}.".format(highest, lowest))
```

### Task 4. What was the largest change in any one day (based on High and Low price)?


```python
highest_price = []
lowest_price = []
change_price = []
for row in json_2017['dataset']['data']:
    highest_price.append(row[2])
    lowest_price.append(row[3])
    change_price.append(row[5])
```

```python
print(change_price)
```

##### The change price columns seems to have all Null values, expect one row. As a result, we cannot use the change column to determine the largest change in one day. Largest change has to be calculated from the high and low prices.

```python
largest_change = [i - j for i, j in zip(highest_price , lowest_price)]
```

```python
print("The largest change in price in a given day is ${}.".format(round(max(largest_change),2)))
```

### Task 5. What was the largest change between any two days (based on Closing Price)?


```python
closing_price = []
for row in json_2017['dataset']['data']:
    closing_price.append(row[4])
```

```python
## Calculating largest change in closing price between two consecutive days
delta_closing = [abs(y - x) for x,y in zip(closing_price,closing_price[1:])]
```

```python
print("The largest change in closing price in any two days is ${}.".format(round(max(delta_closing),2)))
```

### Task 6. What was the average daily trading volume during this year?


```python
trading_vol = []
for row in json_2017['dataset']['data']:
    trading_vol.append(row[6])
```

```python
avg_trading_vol = sum(trading_vol)/len(trading_vol)
```

```python
print("The average daily trading volume in 2017 is {} shares.".format(round(avg_trading_vol,2)))
```

### Task 7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
def median(x):
    x = sorted(x)
    list_len = len(x) 
    num = list_len//2
    if list_len%2==0:
        middlenum = (x[num]+x[num-1])/2
    else:
        middlenum = x[num]
    return middlenum
```

```python
med = median(trading_vol)
```

```python
print("The median trading volume in 2017 is {} shares.".format(round(med,2)))
```
