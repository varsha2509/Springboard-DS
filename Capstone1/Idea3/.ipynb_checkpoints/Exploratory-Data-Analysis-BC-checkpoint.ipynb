```python
BC_df.describe()
```

#### There are 21,488 measurements taken in the Oakland area and the average concentration of Black Carbon between June 2015 - May 2016 is 0.72 ug/m3. 
#### This is roughly the average state-wide concentrations measured back in 2004 - 2005 indicating that the black carbon concentrations are pretty high in the area. The maximum concentration is as high as 8.1 ug/m3 in some areas, indicating that most of the concentration must be from large industrial processes or diesel fueled vehicles. 
##### The main sources of black carbon are from diesel-fueled vehicles, industrial processes, residential fireplaces and woodstoves. Black Carbon is a component of a type of PM2.5.



```python
NO2_df.describe()
```

#### Similar to black carbon, the average NO2 concentration in the area is 17 ppb, which is lower than the National Ambient Air quality Standards of 53 ppb and California Ambient Air Quality Standards of 30 ppb. However, the maximum concentration in the area is 60 ppb, indicating the NO2 concentrations in some areas exceed standards. This could be due to proximity of measurement areas of traffic or industrial sources.

```python
BC_Facility = pd.read_csv("Data/BC_Facility_EmsDist_Inputs.csv")
```

```python
BC_Facility.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
BC_Facility_melt = BC_Facility.melt(id_vars = ['BC Value'])
```

```python
BC_Facility_melt.head(5)
```

```python
BC_Facility_melt['value'].describe()
```

#### Filtering only for facilities having an emissions/distance value > 50th percentile. 

```python
lower, upper = BC_Facility_melt['value'].quantile([0.5,1])
BC_Facility_melt_query = BC_Facility_melt.query('{low}<value<{high}'.format(low=lower,high=upper))
```

```python
BC_Facility_melt_query['value'].describe()
```

```python
NO2_Facility = pd.read_csv("Data/NO2_Facility_EmsDist_Inputs.csv")
```

```python
NO2_Facility.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
NO2_Facility_melt = NO2_Facility.melt(id_vars = ['NO2 Value'])
```

```python
NO2_Facility_melt.head()
```

```python
NO2_Facility_melt.sort_values(['NO2 Value','value'], ascending=[True, False])
```

### <font color='red'>  Is the black carbon concentration higher when there are facilities close by that have high PM2.5 emissions? </font>

```python
ax = sns.scatterplot(x='value', y="BC Value", data=BC_Facility_melt_query, s=16)
plt.xlabel('Emissions (tons)/Distance (mi)', fontsize = 14)
plt.ylabel('BC Value (ug/m3)', fontsize = 14)
```

```python
ax = sns.scatterplot(x='value', y="NO2 Value", data=NO2_Facility_melt, s=18)
plt.xlabel('Emissions (tons)/Distance (mi)', fontsize = 14)
plt.ylabel('NO2 Concentration (ppb)', fontsize = 14)
```

```python
### Identify the type of facilities that have an emissions/distance between 10 - 50
BC_Facility_melt_query[(BC_Facility_melt_query['value']>10) & (BC_Facility_melt_query['value']<50)]['variable'].unique()
```

```python
# Estimate the average concentration of black carbon for facilities with emissions/distance between 10 - 50
BC_Facility_melt_query[(BC_Facility_melt_query['value']>10) & (BC_Facility_melt_query['value']<50)]['BC Value'].mean()
```

###### The above plot and calculations indicate that facilities with an emissions/distance term between 10 - 50 result in an average Black Carbon concentration of 0.90 ug/m3, which is higher than the mean black carbon concentration from all measurements in the area. This could indicated that facilities with high emissions, or facilities close to measurement locations tend to contribute to higher black carbon concentrations. Some of the specific sites that contribute to high concentrations include manufacturing facilities, foundaries, food production plants, electricity generating units and asphalt and concerete plants
